```python
import os
import pandas as pd
import json

# Set the path to the folder containing the JSON files
results_folder = "./results"

# Initialize an empty list to store the individual DataFrames
dataframes = []

# Iterate over all files in the results folder
for file in os.listdir(results_folder):
    # Check if the file is a JSON file
    if file.endswith(".json"):
        # Load the JSON file
        file_path = os.path.join(results_folder, file)
        with open(file_path, "r") as f:
            data = json.load(f)
        
        # Convert the JSON data to a Pandas DataFrame and add it to the list
        df = pd.DataFrame(data, index=[0])
        #df = pd.json_normalize(data)
        dataframes.append(df)

# Combine all individual DataFrames into a single DataFrame
combined_df = pd.concat(dataframes, ignore_index=True)
```
